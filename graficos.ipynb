```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df_agrupado_por_dia =df.groupby('day', sort=False).agg({'total_bill' :np.sum}).reset_index()
df_agrupado_por_dia
```

```python
px.bar(df_agrupado_por_dia, x='day', y='total_bill')
```

```python
df_agrupado_por_dia_e_sexo = df.groupby(['day','sex'], sort=False).agg({'total_bill':np.sum}).reset_index()
df_agrupado_por_dia_e_sexo
px.bar(df_agrupado_por_dia_e_sexo, x='day', y='total_bill', color='sex', barmode='group')
```

```python
px.scatter_3d(df, x='total_bill', y='tip', z='size', color='day', hover_name='time')
```

```python
quantidade_pedidos_por_dia= df['day'].value_counts()
quantidade_pedidos_por_dia

```

```python
quantidade_pedidos_por_dia.index
```

```python
quantidade_pedidos_por_dia.values
```

```python
px.pie(names=quantidade_pedidos_por_dia.index, values=quantidade_pedidos_por_dia.values)
```

```python
quantidade_pedidos_por_sex= df['sex'].value_counts()
quantidade_pedidos_por_sex
```

```python
px.pie(names=quantidade_pedidos_por_sex.index, values=quantidade_pedidos_por_sex.values)
```
